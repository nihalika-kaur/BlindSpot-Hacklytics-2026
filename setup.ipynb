```python
import kagglehub
import shutil
import os

# 1. Get the path where Kaggle hid the files
cache_path = kagglehub.dataset_download("andrewmvd/ocular-disease-recognition-odir5k")
print(f"Files are currently at: {cache_path}")

# 2. Define your local target
local_data_path = "./data"
if not os.path.exists(local_data_path):
    os.makedirs(local_data_path)

# 3. Manual Move: Copy everything from cache to local data
for item in os.listdir(cache_path):
    s = os.path.join(cache_path, item)
    d = os.path.join(local_data_path, item)
    if os.path.isdir(s):
        if os.path.exists(d):
            shutil.rmtree(d) # Clear old version if it exists
        shutil.copytree(s, d)
    else:
        shutil.copy2(s, d)

print(f"✅ Success! Check your local folder: {os.path.abspath(local_data_path)}")
```
